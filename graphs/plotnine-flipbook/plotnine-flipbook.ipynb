The [ggplot](https://ggplot2.tidyverse.org/) library is a great way to make attractive graphics, but unfortunately it's an R library, so we can't use it with Python.

Instead, we'll use [plotnine](https://github.com/has2k1/plotnine), which is an attempt to bring ggplot to Python.

```python
from plotnine import *
```

```python
import pandas as pd
```

```python
df = pd.read_csv("christmas tree sales.csv")
df.head()
```

```python
ggplot(df)
```

```python
ggplot(df) + aes(x='Year')
```

```python
(
    ggplot(df)
    + aes(x='Year')
)
```

```python
(
    ggplot(df)
    + aes(x='Year')
    + aes(y='Trees sold (millions)')
)
```

```python
(
    ggplot(df)
    + aes(x='Year')
    + aes(y='Trees sold (millions)')
    + geom_point()
)
```

```python
(
    ggplot(df)
    + aes(x='Year')
    + aes(y='Trees sold (millions)')
    + geom_point()
    + aes(color='Type of tree')
)
```
